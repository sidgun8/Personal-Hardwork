```python
nums = [1,5, 9,2, 3, 4, 6, 8, 11]
target = 10


def twosum(nums,target):

    remains = {}

    for i, x in enumerate(nums):
        complement = target - x

        if complement in remains:
            return remains[complement], i
        
        remains[x]=i


print(twosum(nums,target))

```

```python
def maxProfit(prices):

    min_price = prices[0]
    max_profit = 0

    for price in prices:

        # Update minimum
        min_price = min(min_price, price)

        # Calculate today's profit
        profit = price - min_price

        # Update maximum
        max_profit = max(max_profit,profit)

    return max_profit

prices = [7,4,5,3,6,12]

maxProfit(prices)
```

```python
nums = [0,3,7,2,5,8,4,6,0,1]


def exists(nums, target):

    for x in nums:
        if x == target:
            return True
        
    return False
    


def longestConsecutive(nums):

    best = 0

    for start in nums:
        current = start
        length = 1

        while exists(nums, current+1) is True:
            current +=1
            length += 1
    
        best = max(best, length)

    return best

nums = [1,2,3,4,5,6,7]

longestConsecutive(nums)
```

```python
def longestConsecutive(nums):

    s = set(nums)

    best = 0

    for num in s:


        if num - 1 not in s:

            current = num
            length = 1

            while current + 1 in s:

                current += 1
                length += 1

            best = max(best, length)

    return best

nums = [1,2,3,4,5,6,7]

longestConsecutive(nums)
```
