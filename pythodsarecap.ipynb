```python
def maxProfit(prices):

    min_price = prices[0]
    max_profit = 0

    for price in prices:

        # Update minimum
        min_price = min(min_price, price)

        # Calculate today's profit
        profit = price - min_price

        # Update maximum
        max_profit = max(max_profit,profit)

    return max_profit

prices = [7,4,5,3,6,12]

maxProfit(prices)
```

```python
def isIsomorphic(s, t):
    mp1 = {}
    mp2 = {}

    if len(s) != len(t):
        return False

    for i in range(len(s)):
        if s[i] in mp1:
            if  mp1[s[i]] != t[i]:
                return False

        elif t[i] in mp2:
            if mp2[t[i]] != s[i]:
                return False
            
        else:
            mp1[s[i]] = t[i]
            mp2[t[i]] = s[i]
    
    return True


s = "rat"
t = "ddd"

isIsomorphic(s,t)



```
