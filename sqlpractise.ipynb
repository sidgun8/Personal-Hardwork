```python
# Pattern 3 — Correlated Subquery

# THIS is the BIG one.

# Interviewers LOVE this.

# Mental Model

# Normal subquery:

# executes ONCE

# Correlated subquery:

# executes FOR EVERY ROW

# This is HUGE.

# Problem

# Find employees earning more than their department average.


# THIS is an ELITE SQL Pattern

# Used for:

# outperforming peers
# anomaly detection
# customer segmentation
# benchmark comparisons
```

```python
%%sql

SELECT e.emp_name,
       e.department,
       e.salary
FROM employee_manager e
WHERE e.salary > (
    SELECT AVG(salary)
    FROM employee_manager
    WHERE department = e.department
);
```

```python
# Mental Model

# EXISTS checks:

# Does at least one row exist?

# Returns:

# TRUE
# FALSE
# Why SELECT 1?

# We don’t care about actual values.

# Only existence.
```

```python
%%sql

SELECT emp_name
FROM employee_manager m
WHERE EXISTS (
    SELECT 1
    FROM employee_manager e
    WHERE e.manager_id = m.emp_id
);
```

```python
# SQL Window Functions Cheat Sheet
# General Syntax
# function_name() OVER (
#     PARTITION BY column
#     ORDER BY column
# )
# 1. ROW_NUMBER()

# Assigns unique row numbers.

# SELECT *,
#        ROW_NUMBER() OVER (
#            PARTITION BY department
#            ORDER BY salary DESC
#        ) AS rn
# FROM employees;
# 2. RANK()

# Same rank for ties, skips numbers.

# SELECT *,
#        RANK() OVER (
#            PARTITION BY department
#            ORDER BY salary DESC
#        ) AS rnk
# FROM employees;
# 3. DENSE_RANK()

# Same rank for ties, no gaps.

# SELECT *,
#        DENSE_RANK() OVER (
#            PARTITION BY department
#            ORDER BY salary DESC
#        ) AS drnk
# FROM employees;
# 4. NTILE()

# Divide rows into buckets.

# SELECT *,
#        NTILE(4) OVER (
#            ORDER BY salary DESC
#        ) AS quartile
# FROM employees;
# 5. SUM() OVER()

# Running totals / partition totals.

# Running Total
# SELECT *,
#        SUM(salary) OVER (
#            ORDER BY salary
#        ) AS running_total
# FROM employees;
# Department Total
# SELECT *,
#        SUM(salary) OVER (
#            PARTITION BY department
#        ) AS dept_total
# FROM employees;
# 6. AVG() OVER()

# Window averages.

# SELECT *,
#        AVG(salary) OVER (
#            PARTITION BY department
#        ) AS dept_avg
# FROM employees;
# 7. COUNT() OVER()

# Window counts.

# SELECT *,
#        COUNT(*) OVER (
#            PARTITION BY department
#        ) AS dept_count
# FROM employees;
# 8. MIN() OVER()

# Window minimum.

# SELECT *,
#        MIN(salary) OVER (
#            PARTITION BY department
#        ) AS dept_min
# FROM employees;
# 9. MAX() OVER()

# Window maximum.

# SELECT *,
#        MAX(salary) OVER (
#            PARTITION BY department
#        ) AS dept_max
# FROM employees;
# 10. LAG()

# Previous row value.

# SELECT *,
#        LAG(salary) OVER (
#            ORDER BY salary
#        ) AS previous_salary
# FROM employees;
# LAG with Offset
# SELECT *,
#        LAG(salary, 2) OVER (
#            ORDER BY salary
#        ) AS two_rows_back
# FROM employees;
# LAG with Default Value
# SELECT *,
#        LAG(salary, 1, 0) OVER (
#            ORDER BY salary
#        ) AS previous_salary
# FROM employees;
# 11. LEAD()

# Next row value.

# SELECT *,
#        LEAD(salary) OVER (
#            ORDER BY salary
#        ) AS next_salary
# FROM employees;
# LEAD with Offset
# SELECT *,
#        LEAD(salary, 2) OVER (
#            ORDER BY salary
#        ) AS two_rows_forward
# FROM employees;
# 12. FIRST_VALUE()

# First value inside partition.

# SELECT *,
#        FIRST_VALUE(salary) OVER (
#            PARTITION BY department
#            ORDER BY salary DESC
#        ) AS highest_salary
# FROM employees;
# 13. LAST_VALUE()

# Last value inside partition.

# SELECT *,
#        LAST_VALUE(salary) OVER (
#            PARTITION BY department
#            ORDER BY salary
#            ROWS BETWEEN UNBOUNDED PRECEDING
#            AND UNBOUNDED FOLLOWING
#        ) AS highest_salary
# FROM employees;
# IMPORTANT

# Without:

# ROWS BETWEEN UNBOUNDED PRECEDING
# AND UNBOUNDED FOLLOWING

# LAST_VALUE often behaves unexpectedly.

# 14. NTH_VALUE()

# Get nth row value.

# SELECT *,
#        NTH_VALUE(salary, 2) OVER (
#            PARTITION BY department
#            ORDER BY salary DESC
#            ROWS BETWEEN UNBOUNDED PRECEDING
#            AND UNBOUNDED FOLLOWING
#        ) AS second_highest
# FROM employees;
# 15. CUME_DIST()

# Cumulative distribution.

# SELECT *,
#        CUME_DIST() OVER (
#            ORDER BY salary
#        ) AS cumulative_distribution
# FROM employees;
# 16. PERCENT_RANK()

# Relative ranking percentage.

# SELECT *,
#        PERCENT_RANK() OVER (
#            ORDER BY salary
#        ) AS percent_rank
# FROM employees;
# 17. Window Frame Clauses

# VERY IMPORTANT advanced topic.

# Running Window
# ROWS BETWEEN UNBOUNDED PRECEDING
# AND CURRENT ROW
# Full Partition
# ROWS BETWEEN UNBOUNDED PRECEDING
# AND UNBOUNDED FOLLOWING
# Sliding Window

# Previous 2 rows:

# ROWS BETWEEN 2 PRECEDING
# AND CURRENT ROW
# Moving Average Example
# SELECT *,
#        AVG(salary) OVER (
#            ORDER BY salary
#            ROWS BETWEEN 2 PRECEDING
#            AND CURRENT ROW
#        ) AS moving_avg
# FROM employees;
# MOST IMPORTANT Interview Patterns
# Top N Per Group
# SELECT *
# FROM (
#     SELECT *,
#            ROW_NUMBER() OVER (
#                PARTITION BY department
#                ORDER BY salary DESC
#            ) AS rn
#     FROM employees
# ) t
# WHERE rn <= 3;
# Second Highest Salary
# SELECT *
# FROM (
#     SELECT *,
#            DENSE_RANK() OVER (
#                ORDER BY salary DESC
#            ) AS rnk
#     FROM employees
# ) t
# WHERE rnk = 2;
# Running Total
# SELECT *,
#        SUM(salary) OVER (
#            ORDER BY salary
#        ) AS running_total
# FROM employees;
# Compare Current vs Previous
# SELECT *,
#        salary - LAG(salary) OVER (
#            ORDER BY salary
#        ) AS diff
# FROM employees;
# Department Average Alongside Rows
# SELECT *,
#        AVG(salary) OVER (
#            PARTITION BY department
#        ) AS dept_avg
# FROM employees;
# Percent Contribution
# SELECT *,
#        salary * 100.0 /
#        SUM(salary) OVER (
#            PARTITION BY department
#        ) AS pct_contribution
# FROM employees;
# SQL Execution Reminder

# Window functions execute AFTER:

# FROM
# JOIN
# WHERE
# GROUP BY
# HAVING

# but BEFORE:

# ORDER BY
# LIMIT
# MOST IMPORTANT Functions To Master

# Priority order:

# ROW_NUMBER()
# RANK()
# DENSE_RANK()
# LAG()
# LEAD()
# SUM() OVER()
# AVG() OVER()
# FIRST_VALUE()

# These dominate interviews.
```

```python
%%sql

with avg_salary_cte as (
    select avg(salary) as avg_salary
    from employee_manager
)


select emp_name, salary
from employee_manager
where salary > (
    select avg_salary
    from avg_salary_cte
)
```

```python
# IMPORTANT

# CTEs are NOT permanent tables.

# They exist:

# only for query duration
# HUGE Advantage

# CTEs improve:

# readability
# debugging
# modularity

# This matters heavily in:

# analytics engineering
# ETL pipelines
# dbt
# Palantir-style workflows
# production SQL
```

```python
%%sql

WITH dept_avg AS (

    SELECT department,
           AVG(salary) AS dept_avg_salary
    FROM employee_manager
    GROUP BY department

),

company_avg AS (

    SELECT AVG(salary) AS company_avg_salary
    FROM employee_manager

)

SELECT d.department,
       d.dept_avg_salary
FROM dept_avg d
CROSS JOIN company_avg c
WHERE d.dept_avg_salary > c.company_avg_salary;
```

```python
%%sql

with ranked_employees as(
    select emp_name, department, salary, row_number() over(
        partition by department
        order by salary desc
    ) as rn
    from employee_manager
)

select *
from ranked_employees
where rn = 1
```

```python
%%sql

WITH dept_salary AS (

    SELECT department,
           SUM(salary) AS total_salary
    FROM employee_manager
    GROUP BY department

),

ranked_departments AS (

    SELECT *,
           RANK() OVER (
               ORDER BY total_salary DESC
           ) AS dept_rank
    FROM dept_salary

)

SELECT *
FROM ranked_departments;
```

```python
# Pattern 4 — Recursive CTEs

# VERY important.

# Used for:

# hierarchies
# trees
# graph traversal
# org structures
# Syntax
# WITH RECURSIVE cte AS (

#     base_case

#     UNION ALL

#     recursive_case

# )

# SELECT *
# FROM cte;
```

```python
%%sql

WITH RECURSIVE hierarchy AS (

    SELECT emp_id,
           emp_name,
           manager_id,
           1 AS level
    FROM employee_manager
    WHERE manager_id IS NULL

    UNION ALL

    SELECT e.emp_id,
           e.emp_name,
           e.manager_id,
           h.level + 1

    FROM employee_manager e
    JOIN hierarchy h
    ON e.manager_id = h.emp_id

)

SELECT *
FROM hierarchy;
```

```python
%%sql

with dept_cum as(
    select distinct department, sum(salary) over(
        partition by department
    ) as cum_dept_sum_salary,
    sum(salary) over() as tot_salary
from employee_manager
)


select department, 100* cum_dept_sum_salary/tot_salary as percentage_contri_dept
from dept_cum
where percentage_contri_dept > 0




```

```python
%%sql

with dept_cum as(
    select distinct department, sum(salary) over(
        partition by department
    ) as cum_dept_sum_salary
from employee_manager
)

select department, row_number() over(
    order by cum_dept_sum_salary desc
) as dept_rank
from dept_cum


```

```python
%%sql

WITH manager_table AS (

    SELECT e.emp_name AS employee,
           m.emp_name AS manager,
           e.salary AS employee_salary,
           m.salary AS manager_salary,
           e.department AS emp_dept,
           m.department AS mng_dept

    FROM employee_manager e
    JOIN employee_manager m
    ON m.emp_id = e.manager_id
),

team_sals AS (

    SELECT manager,

           SUM(employee_salary) OVER (
               PARTITION BY manager
           ) AS team_sal_sum,

           AVG(employee_salary) OVER () AS company_avg_salary

    FROM manager_table
)

SELECT DISTINCT manager
FROM team_sals
WHERE team_sal_sum > company_avg_salary;
```

```python
# Now we practise recursive queries; these are the initially confusing but eventually very powerful part of SQL; its syntax goes like

# WITH RECURSIVE cte_name AS (

#     -- BASE CASE

#     initial query

#     UNION ALL

#     -- RECURSIVE STEP

#     query referencing cte_name

# )

# SELECT *
# FROM cte_name;
```

```python
%%sql

WITH RECURSIVE hierarchy AS (

    -- Base Case
    SELECT emp_id,
           emp_name,
           manager_id,
           1 AS level

    FROM employee_manager
    WHERE manager_id IS NULL

    UNION ALL

    -- Recursive Step
    SELECT e.emp_id,
           e.emp_name,
           e.manager_id,
           h.level + 1

    FROM employee_manager e
    JOIN hierarchy h
    ON e.manager_id = h.emp_id

)

SELECT MAX(level) AS deepest_level
FROM hierarchy;
```
