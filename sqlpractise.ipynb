```python
# Lets practise subqueries now; they are more important

# | Type                | Meaning                     |
# | ------------------- | --------------------------- |
# | Scalar subquery     | returns one value           |
# | IN subquery         | returns list                |
# | Correlated subquery | depends on outer row        |
# | EXISTS              | checks existence            |
# | Derived table       | subquery as temporary table |
```

```python
%%sql

select emp_name, salary, department
from employee_manager
where department = (select distinct department
from employee_manager
where salary > 100000)
```

```python
# Pattern 3 — Correlated Subquery

# THIS is the BIG one.

# Interviewers LOVE this.

# Mental Model

# Normal subquery:

# executes ONCE

# Correlated subquery:

# executes FOR EVERY ROW

# This is HUGE.

# Problem

# Find employees earning more than their department average.
```

```python
%%sql

SELECT e.emp_name,
       e.department,
       e.salary
FROM employee_manager e
WHERE e.salary > (
    SELECT AVG(salary)
    FROM employee_manager
    WHERE department = e.department
);
```
