```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data
df = yf.download("AAPL MSFT", start="2025-01-01", end="2025-12-31")['Close']

# Normalize prices to a base of 100
df_growth = (df / df.iloc[0]) * 100

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df_growth.index, df_growth['AAPL'], label='AAPL')
plt.plot(df_growth.index, df_growth['MSFT'], label='MSFT')
plt.title('Normalized Investment Growth Comparison')
plt.xlabel('Date')
plt.ylabel('Value of $100 Investment')
plt.grid(True)
plt.legend()
plt.show()
```
