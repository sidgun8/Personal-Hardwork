```python
# 2 pointers pattern, this is to reverse array

def reverseArr(arr):
    left,right=0,len(arr)-1
    while left < right:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1
    return arr

marks = [1,2,3,5,6]

reverseArr(marks)

```

```python
# find max

def findMax(arr):
    max = arr[0]
    for i in range(len(arr)):
        if arr[i]> max:
            max = arr[i]
    return max

arr = [6,7,3,12,15,6,2]

findMax(arr)
```

```python
def subarraysum(arr,k):
    mp = {0:1}
    curr_sum = 0
    count = 0

    for x in arr:
        curr_sum += x

        if curr_sum - k in mp:
            count += mp[curr_sum - k]

        mp[curr_sum] = mp.get(curr_sum,0)+1

    return count

 
subarraysum(arr,12)

```

```python
# from here we start doing questions about SLIDING WINDOWS

# SLiding windows are useful as we maintain a window of [left,right] and move it carefully; the core idea is
# Expand window when valid and shrink window when invalid

# Its template looks like

def sliding_window(arr,k):
    left = 0
    curr = 0   #can be sum / count / etc
    result = 0 # Can be min/max/count

    for right in range(len(arr)):
        # Expanding window
        curr += arr[right]

        # Shrinking wundow until valid
        while curr > k:
            curr -= arr[left]
            left +=1

        # Update your answer
        result = max(result, right - left + 1)

    return result
```

```python
# ML Engineering style OOP examples

class MLModel:
    def __init__(self,model_name):
        self.model_name = model_name
        self.is_trained = False

    def train(self,data):
        print(f"Training {self.model_name} on {len(data)} samples.")
        self.is_trained =True

    def predict(self,x):
        if not self.is_trained:
            print("Model not trained")
            return
        return f"Prediction for {x}"
    
model = MLModel("RandomForest")

#model.train([1,2,3])

print(model.is_trained)

print(model.predict(5))
```

```python
def backspstrcomp(s1,s2):
    stack1 = []
    stack2 = []

    for i in s1:
        if i != "#":
            stack1.append(i)
        else:
            if stack1:
                stack1.pop()

    for i in s2:
        if i != '#':
            stack2.append(i)
        else:
            if stack2:
                stack2.pop()        

    return stack1 == stack2


s1 = "##c"
s2 = "addd#c"

backspstrcomp(s1,s2)

```

```python
def dailyTemperatures(temps):

    stack = []
    result = [0] * len(temps)        #this is very nice and important you will get a array for result that looks like if len(tempts)=4 then result is [0,0,0,0]

    for i in range(len(temps)):

        while stack and temps[i] > temps[stack[-1]]:

            prev_index = stack.pop()

            result[prev_index] = i - prev_index

        stack.append(i)

    return result

temps = [73,74,75,71,69,72,76,73]

dailyTemperatures(temps)

```

```python
# This is a problem on setting up of min stack

# this is very important as you learn to trade memory for speed, here we log the min regularly; and hence we go from O(n) to O(1)

class MinStack:

    def __init__(self):
        self.stack = []
        self.min_stack = []


    def push(self,val):
        self.stack.append(val)

        if not self.min_stack:
            self.min_stack.append(val)

        else:
            self.min_stack.append(
                min(val,self.min_stack[-1])
            )

    def pop(self):
        if self.stack:
            self.stack.pop()
            self.min_stack.pop()

    def top(self):
        return self.stack[-1]
    
    def getmin(self):
        return self.min_stack[-1]
    

obj = MinStack()

obj.push(3)
obj.push(1)
obj.top()
print(obj.stack)
print(obj.top())
```

```python
def largestRectangleArea(heights):

    stack = []   # stores indices
    max_area = 0

    heights.append(0)   # sentinel

    for i in range(len(heights)):

        while stack and heights[i] < heights[stack[-1]]:

            h = heights[stack.pop()]

            if not stack:
                width = i
            else:
                width = i - stack[-1] - 1

            area = h * width

            max_area = max(max_area, area)

        stack.append(i)

    return max_area


heights = [2,5,1,6,3,4]
largestRectangleArea(heights)

```

```python
from collections import deque

class MovingAverages:
    def __init__(self,size):
        self.size = size
        self.q = deque()
        self.window_sum = 0

    
    def next(self,val):

        self.q.append(val)
        self.window_sum += val

        if len(self.q) > self.size:
            removed = self.q.popleft()
            self.window_sum -= removed

        return self.window_sum / len(self.q)
    

q = MovingAverages(2)

q.next(3)
q.next(5)
#q.next(7)


```

```python
def reverseKGroups(head,k):
    dummy = Node(0)
    dummy.next = head

    group_prev = dummy

    while True:

        # finding the kth Node
        kth = group_prev

        for i in range(k):
            kth = kth.next

            if not kth:
                return dummy.next
            
        group_next = kth.next

        #reverse group
        prev = group_next
        curr = group_prev.next


        while curr != group_next:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp

        # reconnecting the grouped nodes
        temp = group_prev.next
        group_prev.next = kth
        group_prev = temp
```

```python
# Lets practise search and sorting algorithms; we start with binary search


# This is binary search
def searchtar(arr,target):
    left = 0
    right = len(arr) - 1

    while left <=right:
        mid = (left+right)//2

        if arr[mid] == target:
            return mid
        
        elif arr[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    
    return -1

        

arr = [1,2,3,4,5,6,7,8,9]
k = 7

index_needed = searchtar(arr,k)

print(index_needed)
```

```python
def searchrotated(arr,target):
    left = 0
    right = len(arr)-1

    while left < right:
        mid = (left+right)//2

        if arr[mid] == target:
            return mid
        
        # checking if left sorted
        if arr[left] < arr[mid]:
            if arr[left] < arr[mid] < target:
                right = mid - 1
            else:
                left = mid + 1

        # checking if right is sorted
        else:
            if arr[mid] < target < arr[right]:
                left = mid + 1

            else:
                right = mid - 1

    return -1

arr = [3,2,1,4,0,4,2,3]
k=0

print(searchrotated(arr,k))

```

```python
# Now we study sorting algorithms

def bubblesort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n-i-1):             # for i = 0, j can take 0,1,2 as itll have range(3)
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1], arr[j]

    return arr

arr = [1,9,5,8]

bubblesort(arr)

arr
```

```python
# Selection sort

# the idea here is to repeteadly find the minimum and push it to the begenning of the array

def selectionsort(arr):
    n = len(arr)

    for i in range(n):
        min_idx =  i

        for j in range(i+1,n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr


arr = [7,6,9,2]

selectionsort(arr)

arr


```
