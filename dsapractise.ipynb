```python
# 2 pointers pattern, this is to reverse array

def reverseArr(arr):
    left,right=0,len(arr)-1
    while left < right:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1
    return arr

marks = [1,2,3,5,6]

reverseArr(marks)

```

```python
# find max

def findMax(arr):
    max = arr[0]
    for i in range(len(arr)):
        if arr[i]> max:
            max = arr[i]
    return max

arr = [6,7,3,12,15,6,2]

findMax(arr)
```

```python
def subarraysum(arr,k):
    mp = {0:1}
    curr_sum = 0
    count = 0

    for x in arr:
        curr_sum += x

        if curr_sum - k in mp:
            count += mp[curr_sum - k]

        mp[curr_sum] = mp.get(curr_sum,0)+1

    return count

 
subarraysum(arr,12)

```

```python
# from here we start doing questions about SLIDING WINDOWS

# SLiding windows are useful as we maintain a window of [left,right] and move it carefully; the core idea is
# Expand window when valid and shrink window when invalid

# Its template looks like

def sliding_window(arr,k):
    left = 0
    curr = 0   #can be sum / count / etc
    result = 0 # Can be min/max/count

    for right in range(len(arr)):
        # Expanding window
        curr += arr[right]

        # Shrinking wundow until valid
        while curr > k:
            curr -= arr[left]
            left +=1

        # Update your answer
        result = max(result, right - left + 1)

    return result
```

```python
# ML Engineering style OOP examples

class MLModel:
    def __init__(self,model_name):
        self.model_name = model_name
        self.is_trained = False

    def train(self,data):
        print(f"Training {self.model_name} on {len(data)} samples.")
        self.is_trained =True

    def predict(self,x):
        if not self.is_trained:
            print("Model not trained")
            return
        return f"Prediction for {x}"
    
model = MLModel("RandomForest")

#model.train([1,2,3])

print(model.is_trained)

print(model.predict(5))
```

```python
def paranbalan(str):

    stack = []

    para = {
        ")": "(",
        "}": "{",
        "]": "["
    }

    for i in str:
        if i in "({[":
            stack.append(i)

        else:
            if not stack:
                return False
            
            if stack[-1] != para[i]:
                return False
            
            stack.pop()

    return len(stack) == 0
    
s= '[()]{'

paranbalan(s)
        
        


# In these stack questions stack[-1] becomes a heavy comparitive parameter; stack[-1] means the last element of the stack ie the top of the stack

        

```

```python
# some more problems on stacks and queues

def remoadjdups(s):
    stack = []

    for i in s:
        if stack and i == stack[-1]:      #stack[-1]
            stack.pop()

        else:
            stack.append(i)

    return "".join(stack)

s = 'abbaca'

remoadjdups(s)
```

```python
def backspstrcomp(s1,s2):
    stack1 = []
    stack2 = []

    for i in s1:
        if i != "#":
            stack1.append(i)
        else:
            if stack1:
                stack1.pop()

    for i in s2:
        if i != '#':
            stack2.append(i)
        else:
            if stack2:
                stack2.pop()        

    return stack1 == stack2


s1 = "##c"
s2 = "addd#c"

backspstrcomp(s1,s2)

```

```python
def nextgreater(arr):
    stack = []
    result = []

    for i in range(len(arr)-1,-1,-1):
        while stack and stack[-1] <= arr[i]:
            stack.pop()


        if not stack:
            result.append(-1)
        else:
            result.append(stack[-1])

        stack.append(arr[i])

    return result[::-1]



arr = [4,2,5]
nextgreater(arr)
```

```python
def dailyTemperatures(temps):

    stack = []
    result = [0] * len(temps)        #this is very nice and important you will get a array for result that looks like if len(tempts)=4 then result is [0,0,0,0]

    for i in range(len(temps)):

        while stack and temps[i] > temps[stack[-1]]:

            prev_index = stack.pop()

            result[prev_index] = i - prev_index

        stack.append(i)

    return result

temps = [73,74,75,71,69,72,76,73]

dailyTemperatures(temps)

```
