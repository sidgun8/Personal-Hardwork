```python
def make_multipliers():
    return [lambda x, i=i: i*x for i in range(4)]

multipliers = make_multipliers()

print([m(2) for m in multipliers])
```

```python
def make_multipliers():
    return [lambda x: i*x for i in range(4)]

multipliers = make_multipliers()

print([m(2) for m in multipliers])
```
