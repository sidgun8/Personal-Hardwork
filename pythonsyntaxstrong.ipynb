```python
def divide_numbers(a, b):
    try:
        print("--- Attempting Division ---")
        result = a / b
    except ZeroDivisionError:
        print("Error: You cannot divide by zero!")
        result = None
    else:
        print("Success! No errors occurred.")
        return result
    finally:
        print("Cleaning up... This line always runs.")
```

```python
divide_numbers(10,0)
```

```python
divide_numbers(5,2)
```

```python
# BAD PRACTICE
try:
    # do something
except:
    print("Something went wrong.")



# Good practise

try:
    file = open("data.txt", "r")
    number = int(file.read().strip())
    result = 100 / number

except FileNotFoundError:
    print("The file data.txt does not exist.")

except ValueError:
    print("The file didn't contain a valid number.")

except ZeroDivisionError:
    print("The file contained a 0, cannot divide by zero.")

except Exception as e:
    # This catches ANY other standard error we didn't think of
    print(f"An unexpected error occurred: {e}")
```
